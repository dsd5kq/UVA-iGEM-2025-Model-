```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('../outputs/simR_0.npz')

# Access arrays:
times = data["times"]
R = data["R"]
R_MRNA = data["R_MRNA"]
R_2 = data["R_2"]
R_4 = data["R_4"]
R_8 = data["R_8"]
TF = data["TF"]

plt.figure(figsize=(10, 6))

# Plot mRNA concentrations
#plt.plot(times, A_MRNA, label='CcdA mRNA')
#plt.plot(times, B_MRNA, label='CcdB mRNA')

#Plot protein concentrations
plt.plot(times, TF, label='Transcription Factor')
#plt.plot(times, R_2, label='Dimer')
#plt.plot(times, R_4, label='Tetramer')
plt.plot(times, R_8, label='Octamer')
#plt.plot(times, R, label='Free CcdR')
plt.plot(times, R_MRNA, label='Free CcdR mRNA')

plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('TF and Free CcdR Concentrations Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
